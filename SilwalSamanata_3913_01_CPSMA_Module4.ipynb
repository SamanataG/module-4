Finding the coefficient of a term in a polynomial of the form (x+k)n where n 1 and n N

Applying the formula n!/[(r!)(n-r)! will yield the coefficient of a specific order in a binomial expansion.

````python
import math
n = int(input("Enter the value if n : "))
r = int(input("Enter the value if r : "))
a = math.factorial(n)
b = math.factorial(r)
c = math.factorial(n-r)
k = int(input("Enter the constant : "))
coefficient = ((a)/(b*c))*(k**(n-r))
print("The coefficient of the", r,"th order of (x-3)⁹ is", coefficient)
````
